```python
import pandas as pd


#Определите, какому фильму было выставлено больше всего оценок 5.0.

ratings = pd.read_csv('ratings.csv')
ratings[ratings.rating == 5].groupby('movieId').count().sort_values(by='userId', ascending=False).head(1)
```

```python
#По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) 
#категорий 4, 12 и 21 за период с 2005 по 2010 год. 
#Не учитывайте в расчётах отрицательные значения quantity.

def baltic_states(country):
    if country in ['Lithuania','Latvia', 'Estonia']:
        return True
    return False


power = pd.read_csv('power.csv')
power['baltic_states'] = power.country.apply(baltic_states)

query_str = 'baltic_states == True and category in [4, 12, 21] and quantity > 0 and 2005 < year <2010'
power.query(query_str)['quantity'].sum() 

```

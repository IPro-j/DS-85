```python
'''
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:
оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.
'''


def rating_class(raiting):
    return 'низкий рейтинг' if raiting <= 2 else 'средний рейтинг' if raiting <= 4 else 'высокий рейтинг'


ratings = pd.read_csv('C:\\Users\\Alexey\\Downloads\\ratings.csv')
ratings['class'] = ratings['rating'].apply(rating_class)

ratings.head()

```

```python
'''
Задание 2
Используйте файл keywords.csv.
Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. 
Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. 
Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
'''

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}


keywords = pd.read_csv('C:\\tmp\\Netology\\Python fo data analyse\\ml-latest-small\\ml-latest-small\\keywords.csv')

def geo_class(keyword):
    for i in geo_data:
        for j in geo_data[i]:
            if j in keyword:
                return i
    return('undefined')


keywords['region'] = keywords.keyword.apply(geo_class)
keywords[keywords.region != 'undefined']


```

```python
'''
В переменную years запишите список из всех годов с 1950 по 2010 года.
Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. 
Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте алгоритм:
для каждой строки пройдите по всем годам списка years;
если номер года присутствует в названии фильма, то функция возвращает этот год, как год выпуска;
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’.
Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.
'''

import numpy as np
import pandas as pd

years = np.arange(1950, 2010)

def  production_year(prod_year):
    for y in years:
        if str(y) in prod_year:
            return y
    return 1900

movies = pd.read_csv('C:\\Users\\Alexey\\Downloads\\movies.csv')
ratings = pd.read_csv('C:\\Users\\Alexey\\Downloads\\ratings.csv')

movies['year'] = movies.title.apply(production_year)

movies.merge(ratings[['movieId', 'rating']].groupby('movieId').mean(), on='movieId').sort_values(by=['rating'], ascending=False)


```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


def web_scrap(keywords):
    res_list={}
    for k in keywords:
        habr_page = pd.DataFrame()
        url = f'https://habr.com/ru/search/?q={k.replace(" ", "%20")}&target_type=posts&order=date'
        req = requests.get(url)
        soup = BeautifulSoup(req.text)
        res = soup.find_all('article', class_='tm-articles-list__item')
        for r in res:
            title = r.find('h2').text
            link = r.find('a', class_='tm-title__link').get('href')
            date = r.find('span', class_='tm-article-datetime-published').find('time').get('title').strip()
            row = {'date':date, 'title':title, 'link':link}
            habr_page = pd.concat([habr_page, pd.DataFrame([row])])
            #print(link)
        res_list[k] = habr_page.drop_duplicates(keep=False)
    return(res_list)


w = web_scrap(['python', 'анализ данных'])
w['python']
```

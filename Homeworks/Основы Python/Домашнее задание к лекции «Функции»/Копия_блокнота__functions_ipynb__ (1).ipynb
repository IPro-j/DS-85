```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


#Пользователь по команде "p" может узнать владельца документа по его номеру
def name_document_owner(doc_number):
  for doc in documents:
    if doc['number'] == doc_number[0]:
      print('Владелец документа: %s' % doc['name'])
      return
  print('Документ не найден в базе')


#Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
def document_directory(doc_number):
  for dir in directories:
    if doc_number[0] in directories[dir]:
      if len(doc_number) == 1:
        print('Документ хранится на полке: %s' % dir)
      return dir
  print('Документ не найден в базе')


#Пользователь по команде "l" может увидеть полную информацию по всем документам
def documents_info(parameters):
  print('Текущий список документов:')
  for doc in documents:
    parameters = [doc['number'], 1]
    print('№: %s, тип: %s, владелец: %s, полка хранения: %s' % (doc['number'], doc['type'], doc['name'], document_directory(parameters)))


#Пользователь по команде "ads" может добавить новую полку
def dir_add(dir_num):
  if dir_num[0] in directories:
    print('Такая полка уже существует. Текущий перечень полок: %s.' %  dir_exists())
    return
  directories[dir_num[0]] = []
  print('Полка добавлена. Текущий перечень полок: %s' % dir_exists())


#Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
def dir_delete(dir_num):
  if dir_num[0] not in directories:
    print('Такой полки не существует. Текущий перечень полок: %s.' %  dir_exists())
    return
  if len(directories[dir_num[0]]) > 0:
    print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: %s' % dir_exists())
    return
  del directories[dir_num[0]]
  print('Полка удалена. Текущий перечень полок: %s' % dir_exists())


#текущий список полок
def dir_exists():
  return(', '.join(list(directories.keys())))


# Пользователь по команде "ad" может добавить новый документ в данные
def doc_add(parameters):
  if parameters[3] not in directories:
    print('Такой полки не существует. Добавьте полку командой as.')
    documents_info([])
    return
  documents.append({'type': parameters[1], 'number': parameters[0], 'name': parameters[2]})
  directories[parameters[3]] = parameters[0]
  documents_info([])


# Пользователь по команде "d" может удалить документ из данных
def doc_delete(doc_number):
  for doc in documents:
    if doc['number'] == doc_number[0]:
      documents.remove(doc)
      dir = document_directory(doc_number)
      directories[dir].remove(doc_number[0])
      print('Документ удален')
      documents_info([])
      return
  print('Документ не найден в базе')
  documents_info([])


#Пользователь по команде "m" может переместить документ с полки на полку
def doc_move(parameters):
  if parameters[1] not in directories:
    print('Такой полки не существует. Текущий перечень полок: %s.' %  dir_exists())
    return
  if doc_is(parameters[0]):
    dir = document_directory(parameters)
    directories[dir].remove(parameters[0])
    directories[parameters[1]].append(parameters[0])
    documents_info([])


# наличие документа в базе
def doc_is(doc_num):
  for doc in documents:
    if doc['number'] == doc_num:
      return True
  print('Документ не найден в базе')
  documents_info([])
  return False


#выполнение комманд
def cmd_run(cmd):
  cmd_list ={
    'p' : [['Введите номер документа:',], name_document_owner,],
    's' : [['Введите номер документа:',], document_directory,],
    'l' : [[], documents_info],
    'ads' : [['Введите номер полки:'], dir_add,],
    'ds' : [['Введите номер полки:'], dir_delete,],
    'ad' : [['Введите номер документа:', 'Введите тип документа:', 'Введите владельца документа:', 'Введите полку для хранения:'], doc_add,],
    'd' : [['Введите номер документа:',], doc_delete,],
    'm' : [['Введите номер документа:', 'Введите номер полки:',], doc_move,],

  }

  if cmd not in  cmd_list:
    print('Команда не найдена')
    return

  parameters = []
  for p in cmd_list[cmd][0]:
    parameters.append(input(p))
  cmd_list[cmd][1](parameters)


while 1:
  cmd = input('Введите команду: ')
  if cmd == 'q':
    print('Работа завершена')
    break
  cmd_run(cmd)





```

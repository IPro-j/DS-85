```python
'''
–ó–∞–¥–∞–Ω–∏–µ 1
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–π

2ùë•2ùë¶3+ùë•ùë¶2+7+1ùë•

ùë•2ùë¶+6ùë¶‚àísin(ùë•ùë¶)+cos(ùë•2)
'''

from sympy import *

x,y = symbols('x y')
```

```python
f = 2*x**2*y**3 + 1/x + y**2*x + 7
f
```

```python
f_diff_x = diff(f, x)
f_diff_x
```

```python
f_diff_y = diff(f, y)
f_diff_y
```

```python
f1 = x**2*y - sin(x*y) + cos(x**2) + 6*y
f1
```

```python
diff(f1, x)
```

```python
diff(f1, y)
```

```python
'''
–ó–∞–¥–∞–Ω–∏–µ 2
–í—ã—á–∏—Å–ª–∏—Ç–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç —Ñ—É–Ω–∫—Ü–∏–∏:

2ùë•2ùë¶3+ùë•ùë¶2+7+1ùë•

–≤ —Ç–æ—á–∫–µ 1,2
'''

grad = [diff(f,var) for var in (x,y)]
grad_ = lambdify([x,y], grad)

grad_(1, 2)
```

```python
'''
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç—Ä–µ—Ö–º–µ—Ä–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏

‚àíùë•21ùë•2+2ùë•1ùë•2+ùë•22
'''


from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

```

```python
def fun(x_1, x_2):
    return x_2**2 - x_2*x_1**2 + 2*x_1*x_2

```

```python
x1_ = np.linspace(-5, 5, 100)
x2_ = x1_

z = fun(x1_, x1_)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x1_, x2_, z, label='parametric curve')
```

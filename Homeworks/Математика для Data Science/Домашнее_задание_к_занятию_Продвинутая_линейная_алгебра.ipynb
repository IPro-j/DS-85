```python
'''
Дан объект в
 пространстве

При помощи линейных отображений:

Уменьшить объект в два раза
Повернуть на 130 градусов
Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])
'''

import numpy as np
import matplotlib.pyplot as plt


def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))


A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])

PolyArea(x,y)
```

```python
'''
Уменьшить объект в два раза (плошадь)
'''

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

B1 = [[1/np.sqrt(2), 0, 0], [0, 1/np.sqrt(2), 0], [0, 0, 1]]
B1 = np.array(B1)


A = A @ B1

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])


PolyArea(x,y)
```

```python
'''
Повернуть на 130 градусов
'''


import math

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

B2=[[math.cos(130*math.pi/180), math.sin(130*math.pi/180), 0],
     [-(math.sin(130*math.pi/180)), math.cos(130*math.pi/180), 0],
     [0, 0,1]]
B2 = np.array(B2)


A = A @ B2

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
'''
Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])
'''

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

B3 = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])


A = A @ B3

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])

plt.plot([-200, 200],[-200, 200])
```

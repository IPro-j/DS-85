```python
'''
Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его
и оставить 8 столбцов для дальнейшего изучения:
surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.
'''


import pandas as pd

head = ('surgery?', 'Age', 'Hospital Number',
		'rectal temperature', 'Pulse', 'respiratory rate',
		'temperature of extremities', 'peripheral pulse', 'mucous membranes',
		'capillary refill time', 'pain', 'peristalsis',
		'abdominal distension', 'nasogastric tube', 'nasogastric reflux',
		'nasogastric reflux PH', 'rectal examination - feces', 'abdomen',
		'packed cell volume', 'total protein', 'abdominocentesis appearance',
		'abdomcentesis total protein', 'outcome', 'surgical lesion?',
		'type of lesion1', 'type of lesion2', 'type of lesion3',
		'cp_data')

horse_data = pd.read_csv('horse_data.csv', header=0, names=head, na_values=['?'], usecols=['surgery?', 'Age',
                                                                'rectal temperature', 'Pulse', 'respiratory rate',
                                                                'temperature of extremities', 'pain', 'outcome',])


horse_data.info()
```

```python
'''
Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.
'''

horse_data.head()
```

```python

# значение 9 для столбца Age - ошибочное значение, проставленна вместо 2
horse_data.loc[horse_data['Age']  == 9, 'Age'] = 2
horse_data.loc[horse_data['Age']  == 2]
horse_data.describe()
```

```python
def hd_outlier(col_name):
    q1 = horse_data[col_name].quantile(0.25)
    q3 = horse_data[col_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = horse_data.query(f'`{col_name}` < {lower_bound} or `{col_name}` > {upper_bound}')
    return(remove_outliers)



print('Кол-во выбросов по  rectal temperature ',  len(hd_outlier('rectal temperature')))
print('Кол-во выбросов по  Pulse ',  len(hd_outlier('Pulse')))
print('Кол-во выбросов по  respiratory rate ', len(hd_outlier('respiratory rate')))

```

```python
'''
Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов.
Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик
и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.
'''


horse_data.isna().sum ()


```

```python
horse_data.isna().mean()*100
```

```python
hd_copy = horse_data.copy()

# по одному пропуску в столбцах surgery? и outcome, заполняем модами,
# исходя из того что  моды остальных показатели (кроме pain, отличается на 1)
# совпадают со значениями в строке с пропусками  surgery? и outcome
hd_copy['surgery?'].fillna(hd_copy['surgery?'].mode()[0], inplace=True)
hd_copy['outcome'].fillna(hd_copy['outcome'].mode()[0], inplace=True)

# исходим из описания:   - In general, the more painful, the more likely it is to require surgery
hd_copy['pain'] = hd_copy['pain'].fillna(horse_data.groupby('surgery?')['pain'].transform('median'))

# частота дыхание зависит от возраста, у молодых выше
hd_copy['respiratory rate'] = hd_copy['respiratory rate'].fillna(horse_data.groupby('Age')['respiratory rate'].transform('median'))

#исходим из описания: animals with painful lesions or suffering from circulatory shock
hd_copy['Pulse'] = hd_copy['Pulse'].fillna(horse_data.groupby('pain')['Pulse'].transform('median'))

# допускаем корреляцию  temperature of extremities и rectal temperature
hd_copy['rectal temperature'] = hd_copy['rectal temperature'].fillna(horse_data.groupby(['temperature of extremities'])['rectal temperature'].transform('median'))

# используем корреляцию pain и temperature of extremitie
hd_copy['temperature of extremities'] = hd_copy['temperature of extremities'].fillna(horse_data.groupby('pain')['temperature of extremities'].transform('median'))

# кол-во пропусков сократилось до  67 в 39 (13%) строках, изначально с пропусками было 145 строк
hd_copy.isna().sum()

# убираем строки с пропусками, работаем далее с 87% выборкой из начального дата фрейма
hd_copy.dropna()

```

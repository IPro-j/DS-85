```python
'''
This dataset consists of 16 features containing information about 203 vehicles available for sale in India

Первоначально dataset содержал только марки авто, поле "car_brend" было добавленно при подготовке данных
'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cars = pd.read_csv('https://raw.githubusercontent.com/IPro-j/DS-85_Houmeworks/main/CARS_1.csv',  sep=';', on_bad_lines='skip')
cars.info()
cars.head()
```

```python
cars.describe()
```

```python
# единичные пропуски были заполнены вручную данными из интернета

cars.isna().sum()
```

```python
brend_count = cars.groupby('car_brend')[['car_name']].count().sort_values('car_name', ascending=False)

brend_count = brend_count.reset_index(names=['car_brend'])
brend_count.rename(columns = {'car_name':'car_names'}, inplace = True)

brend_count.head(10).plot.barh(x='car_brend')

'''
ранжирование по кол-ву разных марок/комплектаций на уровень бренда.
возможность поддержания разнообразия на рынке говорит о технологично развитом производстве,
налаженной логистке/(производстве на месте) запчастей;
явные лидеры Мерседес, БМВ и Maruti(индийский бренд Suzuki),
Mahindra(Индийский многонациональный конгломерат со штаб-квартирой в Мумбаи)

'''
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)


transmission_type = cars.groupby('transmission_type')[['car_name']].count().sort_values('car_name', ascending=False)
transmission_type = transmission_type.reset_index(names=['transmission_type'])
transmission_type.plot(ax=axes[0], kind='pie', y='car_name', title = 'Transmission type', legend = 'transmission_type')

fig.legend(transmission_type['transmission_type'])

transmission_type1 = cars.groupby('transmission_type')[['reviews_count']].sum().sort_values('reviews_count', ascending=False)
transmission_type1 = transmission_type1.reset_index(names=['transmission_type'])
transmission_type1.plot(ax=axes[1], kind='pie', y='reviews_count', title = 'Transmission type')

'''
доли представленных автомобилей с разной транссмисией на рынке  плюс доля отзывов(популярность)

автомат - безоговорочный лидер, популярность электрических авто не соответсвует доли марок на рынке
'''
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

brend_rate = cars.groupby('car_brend')[['rating']].mean().sort_values('rating', ascending=False).head(10)
brend_rate.plot.bar(ax=axes[0], ylim =[(brend_rate['rating'].min() - 0.05) , brend_rate['rating'].max()])


brend_rate = cars.groupby('car_brend').agg({'rating' : 'mean', 'reviews_count' : 'sum'}).sort_values('reviews_count', ascending=False).head(10)
brend_rate.reset_index(names=['car_brend'], inplace=True)
brend_rate.sort_values('rating', ascending=False, inplace=True)
brend_rate[['car_brend', 'rating']].plot.bar(ax=axes[1], ylim = [(brend_rate['rating'].min() - 0.05) , brend_rate['rating'].max()], x = 'car_brend')

'''
усреднЁный на уровень бренда рейтинг авто, плюс усредненный рейтинг самых популярных брендов(кол-во отзывов)
единственное пересечение компания Tata(местный производитель, специфика, требует доп. анализа)

без учета популярности(покупаемости) в рейтинге в основном престижные бренды,
с учетом популярности в топах бренды с ориентацией на средний сегмент рынка

'''
```

```python
'''
рейтинг по кол-ву отзывов(популярности), это ко второму графику - см. выше

первая тройка - местные бренды,
выводы - собственное автомобильное производство в Индии достаточно развито и привлекательно для внутренего рынка
'''

brend_reviews = cars.groupby('car_brend')[['reviews_count']].sum().sort_values('reviews_count', ascending=False).head(10)
brend_reviews.reset_index(names=['car_brend'], inplace=True)
brend_reviews.style.bar(align='mid', color='green')
```
